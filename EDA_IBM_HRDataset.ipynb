```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No':0})
```

```python
# Department vs Attrition

df.groupby('Department')['Attrition'].mean().plot(kind='bar')
plt.title("Attrition Rate by Department")
plt.show()
```

### Insight -> The Sales department shows the highest attrition rate, indicating potential issues such as high pressure, targets, or job satisfaction compared to other department

```python
# Salary vs Attrition
df.groupby('Attrition')['MonthlyIncome'].mean()
```

### Insight -> This suggest that improving salary structures or offering compatitve compasation could help to reduce employee turnover

```python
# Job Satisfaction Impact

df.groupby('JobSatisfaction')['Attrition'].mean().plot(kind = 'bar')
plt.title("Attrition vs Job Satisfaction")
plt.show()
```

### Insight -> Employees with low job satisfaction exhibit significantly higher attrition rates, indicating that employee engagement and satisfaction are key factors influencing retentions.

```python
# Years at Company

df.groupby('YearsAtCompany')['Attrition'].mean().plot()
plt.title("Attrition vs Years at Company")
plt.show()
```

### Insight -> Employees are more likely to leave within first few years of joining the company , indicating potential onboarding or early angagement issues.
